# Wuzzuf Job Scraper

A web scraping project that extracts job listings (title, company, location) from [Wuzzuf.net](https://wuzzuf.net), Egypt's leading job search platform.

> ⚠️ **Important:** Wuzzuf's CSS class names are auto-generated (CSS-in-JS) and can change whenever the site is redeployed. If this scraper returns no results, open the page in your browser's DevTools, inspect a job card, and update the selectors in `src/scraper.py`. Always check a website's `robots.txt` and Terms of Service before scraping, and keep delays between requests to be a good citizen.


## 1. Imports

```python
from src.scraper import fetch_page, parse_job_cards, scrape_jobs
from src.cleaning import clean_jobs, make_full_urls

import pandas as pd
```

## 2. Exploring a Single Page

Before scraping many pages, it's good practice to fetch and inspect a single page first to confirm the selectors still match the current site structure.

```python
soup = fetch_page("https://wuzzuf.net/search/jobs?q=AI%20Engineer&a=spbg")
soup.title.text
```

```python
jobs_on_page = parse_job_cards(soup)
pd.DataFrame(jobs_on_page).head()
```

## 3. Scraping Multiple Pages

We loop through several pages of results, with a short delay between requests to avoid overloading the server.

```python
df = scrape_jobs(country="Egypt", max_pages=5)
df.shape
```

## 4. Cleaning the Scraped Data

- Drop duplicate rows (pagination can sometimes overlap)
- Drop rows where parsing failed to find a title
- Convert relative links into full, clickable URLs

```python
df = clean_jobs(df)
df = make_full_urls(df)
df.head()
```

## 5. Saving the Results

```python
df.to_csv("../data/wuzzuf_jobs.csv", index=False)
print(f"Saved {len(df)} job listings.")
```

## 6. Troubleshooting

If `parse_job_cards` returns an empty list:

1. Open the search results page in your browser and inspect a job card with DevTools.
2. Find the current class names for the job card container, title link, company, and location.
3. Update the `SELECTORS` dictionary at the top of `src/scraper.py`.
4. Re-run this notebook.

This is a normal part of maintaining any scraper — site frontends change over time.


## Conclusion

This project scrapes structured job data (title, company, location, link) from Wuzzuf's search results, with basic cleaning and deduplication. The resulting CSV can feed into further analysis — e.g. most in-demand job titles, skill trends by role, or geographic distribution of openings.
